**1 - Simulação de Sinais Estocásticos com $N=2^{12}$ valores de medidas.** 

**1.1 - Utilize o algoritmo powenoise.m e gere os seguintes ruídos $1/f^{β}$:**

S1: $\beta=0$ (white noise)

S2: $\beta=1$ (pink noise)

S3: $\beta=2$ (red noise)

```python
white_noise = powernoise(beta=0, N=2**12)
pink_noise = powernoise(beta=1, N=2**12)
red_noise = powernoise(beta=2, N=2**12)

S1, S2, S3 = white_noise, pink_noise, red_noise

fig = plt.figure(figsize=(15, 10))
fig.subplots_adjust(hspace = .3, wspace = .2)
W = fig.add_subplot(2, 1, 1)
W.set_ylabel('Amplitude')
W.set_xlabel('Time')
W.plot(white_noise, color='#0100FF')
W.set_title(r'White Noise $\beta$=0', fontsize='16')

P = fig.add_subplot(2, 2, 3)   
P.set_ylabel('Amplitude')
P.set_xlabel('Time')
P.plot(pink_noise, color='hotpink')
P.set_title(r'Pink Noise $\beta$=1', fontsize='16')

R = fig.add_subplot(2, 2, 4)   
R.set_ylabel('Amplitude')
R.set_xlabel('Time')
R.plot(red_noise, color='red')
R.set_title(r'Red Noise $\beta$=2', fontsize='16')
```
